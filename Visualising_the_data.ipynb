# Line plot

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_net_mig():
  excel_url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true'
  
  #1. Read sheet_name Country Migration
  country_mig_df = pd.read_excel(excel_url, sheet_name='Country Migration')

  #2. create a new dataframe containing only the five columns holding net migration
  col_list = ['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019']
  mig_2015_2019 = pd.DataFrame()
  mig_2015_2019 = country_mig_df[col_list]
  
  #3. create a new data variable to hold the means of the five columns
  mig_mean = mig_2015_2019.mean()
  #print(mig_mean)

  #4. create a labels variable to hold the keys (column headings)
  labels = mig_2015_2019.keys()
  #print(labels)

  #5. use plt.plot(x, y) to plot your line chart
  plt.plot(labels, mig_mean)
  plt.xticks(rotation=45)
  plt.grid()
  plt.show()

plot_net_mig()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress as lr

def scatterplot_wellbeing_stress():
  csv_url = 'https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true'
  
  #1. Read the dataset
  psych_data = pd.read_csv(csv_url)

  #2. create a variable which groups the means of data by firstlanguage
  df_stress = psych_data['Stress']
  df_wellbeing = psych_data['Wellbeing']

  # Plot the data
  plt.scatter(df_stress, df_wellbeing, color='orange', label='Actual values')
  plt.title('Stress vs wellbeing scatter plot')
  plt.xlabel('Stress')
  plt.ylabel('Wellbeing')
  # If show() is called here, below plot will be a new plot in new figure
  #plt.show()

  # Line of best fit
  regression = lr(df_stress, df_wellbeing)
  print(regression)
  print(f"R-squared: {regression.rvalue**2:.6f}")

  slope = regression.slope
  intercept = regression.intercept
  # y^ = b0 + b1x -> b1=slope, b0=intercept
  # y^ is the estimated value, y- -> mean value, y-> actual value
  # R2 -> sum(y^ - y-)^2/ sum(y - y-)^2
  # R2=1 -> estimated value == actual value -> perfect fit
  plt.plot(df_stress, slope*df_stress + intercept, label='fitted line')
  plt.legend()
  plt.show()

  
scatterplot_wellbeing_stress()
```

# Data story


> Aim is to analyze the World Happiness Report of 2015 and look into the factors that contributed to these rankings.
Analysis show that the top countries that have high ranks on the happiness scale all rank highly on all of the main factors that are found to support happiness: family, health, income, freedom, and economy.

The graphs below show that people have similar level of freedom, generosity and trust in their governments for countries who have ranked higher in happiness score.

The graphs below look at relationship between economy and trust in government in most happy and least happy countries.
Looks like majority of the people do not trust their governments.

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_factors(**kwargs):
  factors = kwargs['factors']
  kwargs['p1'].plot(x='Country', y=factors, kind="bar", figsize=(6,5), stacked=kwargs['stack'])
  plt.legend(loc="lower left",bbox_to_anchor=(0.8,1.0))
  kwargs['p2'].plot(x='Country', y=factors, kind="bar", figsize=(6,5), stacked=kwargs['stack'])
  plt.legend(loc="lower left",bbox_to_anchor=(0.8,1.0))
  plt.show()

def happiness_analysis_2015():
  excel_url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true'
  df_15 = pd.read_excel(excel_url)

  # 1. Sort happiness rank in ascending order
  # 2. Pick top and bottom 10 countries
  # 3. Plot bar graph of each factor for those countries (use separate plots for top and bottom)
  # 4. Narrow down and pick only 5 factors which seem reasonable
  # 5. Check if there is any correlation b/w them

  # Set some display options when run
  pd.options.display.max_rows= 20
  pd.options.display.max_columns= 12
  df_15 = df_15.sort_values(by=['Happiness Rank'], ascending=True)
  # To get an overall idea of numbers
  print(df_15.describe())
  
  # Get top 10 country names in a list
  max_rank = len(df_15) -1
  top_c = df_15.loc[0:9]
  bottom_c = df_15.loc[max_rank-10:max_rank]
  
  # Plot each factor for top and bottom countries
  factors = ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)']
  plot_factors(p1=top_c, p2=bottom_c, factors=factors, stack=True)
  
  factors = ['Freedom', 'Generosity', 'Trust (Government Corruption)']
  plot_factors(p1=top_c, p2=bottom_c, factors=factors, stack=True)
  
  factors = ['Economy (GDP per Capita)', 'Trust (Government Corruption)']
  plot_factors(p1=top_c, p2=bottom_c, factors=factors, stack=False)
  
 

happiness_analysis_2015()
```
